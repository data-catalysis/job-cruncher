```python
import spacy
# Load a larger model with vectors
nlp = spacy.load('en_core_web_md')
```

```python
doc = nlp('Data Scientist')
print(doc.similarity(nlp(onet_oc_df.iloc[0]['titles'])))
print(doc.similarity(nlp(onet_oc_df.iloc[1]['titles'])))
print(doc.similarity(nlp(onet_oc_df.iloc[2]['titles'])))
print(doc.similarity(nlp(onet_oc_df.iloc[3]['titles'])))
```

```python
for i in range(4):
    occupation_name = onet_oc_df.iloc[i]['name']
    alternate_titles = onet_oc_df.iloc[i]['titles'].split(',')
    sim_scores = [doc.similarity(nlp(title)) for title in alternate_titles]
    sorted_indx = np.argsort(sim_scores)[::-1]
    print('{} + {}: {}'.format(doc.text, occupation_name, doc.similarity(nlp(occupation_name))))
    for i,v in enumerate(sorted_indx):
        print('{} - {}: {}'.format(doc.text, alternate_titles[v], sim_scores[v]))
    print('---')
```
